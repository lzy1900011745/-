```python
import numpy as np
```

```python
X_1 = X1.drop([ 'age', 'sex', 'test_time','age_group'],axis=1).values
X_2 = X2.drop([ 'age', 'sex', 'test_time','age_group'],axis=1).values
X_3 = X3.drop([ 'age', 'sex', 'test_time','age_group'],axis=1).values
X_4 = X4.drop([ 'age', 'sex', 'test_time','age_group'],axis=1).values
X_5 = X5.drop([ 'age', 'sex', 'test_time','age_group'],axis=1).values
y1 =y1.values
y2 =y2.values
y3 =y3.values
y4 =y4.values
y5 =y5.values
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

torch.manual_seed(0)

# 将数据集转换为PyTorch张量
X_train_tensor = torch.FloatTensor(np.concatenate([X_2, X_3, X_4], axis=0))
y_train_tensor = torch.FloatTensor(np.concatenate([y2, y3, y4], axis=0).reshape(-1,1))  
X_test1_tensor = torch.FloatTensor(X_1)
y_test1_tensor = torch.FloatTensor(y1.reshape(-1,1))
X_test2_tensor = torch.FloatTensor(X_5)
y_test2_tensor = torch.FloatTensor(y5.reshape(-1,1))

# 创建DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test1_dataset = TensorDataset(X_test1_tensor, y_test1_tensor)
test2_dataset = TensorDataset(X_test2_tensor, y_test2_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test1_loader = DataLoader(test1_dataset, batch_size=32, shuffle=False)
test2_loader = DataLoader(test2_dataset, batch_size=32, shuffle=False)

# 定义神经网络模型
class SimpleNN(nn.Module):
    def __init__(self, input_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(64, 32)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(32, 1)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.fc3(x)
        return x

# 实例化模型、损失函数和优化器
model = SimpleNN(X_train_tensor.shape[1])
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 训练神经网络
epochs = 10
for epoch in range(epochs):
    model.train()
    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

# 评估模型
def evaluate_model(model, X_test, y_test):
    model.eval()
    with torch.no_grad():
        y_pred_tensor = model(X_test)
        y_pred = y_pred_tensor.numpy()
        rmse = mean_squared_error(y_test, y_pred, squared=False)
        return rmse

rmse1 = evaluate_model(model, X_test1_tensor, y_test1_tensor)
print(f'Root Mean Squared Error on the first test set: {rmse1}')

rmse2 = evaluate_model(model, X_test2_tensor, y_test2_tensor)
print(f'Root Mean Squared Error on the second test set: {rmse2}')
```

```python
# 在X2，X3，X4上分别训练一个模型
models = []
for X_train, y_train in [(X_2, y2), (X_3, y3), (X_4, y4)]:
    model = SimpleNN(X_train.shape[1])
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    train_dataset = TensorDataset(torch.FloatTensor(X_train), torch.FloatTensor(y_train))
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    for epoch in range(epochs):
        model.train()
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs.view(-1), targets)
            loss.backward()
            optimizer.step()
    models.append(model)


```

```python
import numpy as np

# 初始化一个3*3的矩阵
matrix = np.zeros((3, 3))

# 对于每一个测试数据
for x in X_5:
    # 计算每个模型的输出
    outputs = [model(torch.FloatTensor(x)).detach().numpy() for model in models]
    # 对于每一对模型
    for i in range(3):
        for j in range(3):
            # 计算f_i(x)*f_j(x)
            matrix[i, j] += outputs[i] * outputs[j]

# 求平均
matrix /= len(X_5)
print(matrix)
```

```python
import cvxpy as cp
import numpy as np

# 假设我们有一个3x3的正定矩阵Gamma
Gamma = np.array(matrix)

# 定义优化变量
q = cp.Variable(3)

q_sou = (len(X_2)/(len(X_2)+len(X_3)+len(X_4)),len(X_3)/(len(X_2)+len(X_3)+len(X_4)),len(X_4)/(len(X_2)+len(X_3)+len(X_4)))
epsilon = 0.5
# 定义优化问题
objective = cp.Minimize(cp.quad_form(q, Gamma))
constraints = [cp.sum(q) == 1, q >= 0,
               cp.norm(q - q_sou) <= epsilon]  
prob = cp.Problem(objective, constraints)

# 求解优化问题
result = prob.solve()

# 输出结果
print("最优值: ", result)
print("最优解: ", q.value)
```

```python
# 定义凸线性组合的权重
tau = 1*result
weights = tau/result * q.value

# 对X1和X5进行预测
for X_test, y_test in [(X_5, y5)]:
    X_test_tensor = torch.FloatTensor(X_test)
    y_test_tensor = torch.FloatTensor(y_test)
    preds = [model(X_test_tensor).detach().numpy() for model in models]
    # 计算凸线性组合
    y_pred = sum(w * pred for w, pred in zip(weights, preds))
    # 计算RMSE
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    print(f'Root Mean Squared Error on the test set: {rmse}')
```
